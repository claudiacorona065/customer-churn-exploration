### Creación de la columna `is_churner`

### Implementación

#### 1. **Preservación de la integridad de los datos:**
No rellenamos los valores nulos en `churn_date` con texto o fechas ficticias. Dejamos los valores como `NaT` para mantener la integridad de los datos.

#### 2. **Creación de la columna `is_churner`:**
La lógica utilizada para crear esta columna es:
- Si `churn_date` no es nulo (`notnull()`), el cliente es un churner (`1`).
- Si `churn_date` es nulo (`NaT`), el cliente no es un churner (`0`).

---

### Explicación
Supongamos que la columna `churn_date` contiene los siguientes valores:

| churn_date        | is_churner |
|--------------------|------------|
| 2023-05-13        | 1          |
| NaT               | 0          |
| 2023-08-20        | 1          |
| NaT               | 0          |
| 2023-06-17        | 1          |

- Los clientes con una fecha en `churn_date` son marcados como **`1`** en `is_churner`, indicando que son churners.
- Los clientes con `NaT` en `churn_date` son marcados como **`0`** en `is_churner`, indicando que están activos.



```python
# Tratamiento de valores nulos
# Para churn_date: marcar explícitamente como "No churn" y crear is_churner
df['churn_date'].fillna(value=pd.NaT, inplace=True)
df['is_churner'] = df['churn_date'].notnull().astype(int)

# Para contract_channel: decidir si eliminar o mantener (aquí lo eliminamos)
df.drop(columns=['contract_channel'], inplace=True)

# Rellenar gender, paymentmethod y zipcode con "Desconocido"
df['gender'].fillna('Desconocido', inplace=True)
df['paymentmethod'].fillna('Desconocido', inplace=True)
df['zipcode'].fillna('Desconocido', inplace=True)

# Verificar si contract_date es eliminable o imputable
missing_contract = df['contract_date'].isnull().sum()
print(f"Faltan fechas de contrato en {missing_contract} registros.")
```

```python
## Crear una copia explícita del DataFrame para evitar problemas con vistas
df = df.copy()

df['paperlessbilling'] = df['paperlessbilling'].astype('category')
df['married'] = df['married'].astype('category')

# Verificar y agregar "Desconocido" 
if 'Desconocido' not in df['paperlessbilling'].cat.categories:
    df['paperlessbilling'] = df['paperlessbilling'].cat.add_categories(['Desconocido'])
if 'Desconocido' not in df['married'].cat.categories:
    df['married'] = df['married'].cat.add_categories(['Desconocido'])


df['paperlessbilling'] = df['paperlessbilling'].fillna('Desconocido')
df['married'] = df['married'].fillna('Desconocido')

print(df.isnull().sum())
```

Con esta imputación, los registros serán integrados en el análisis sin introducir sesgos significativos en las métricas calculadas.

```python
# Detección de outliers usando IQR
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers en {column}:")
    print(outliers[[column]])
    print("\n")

    # Boxplot
    plt.boxplot(df[column].dropna(), vert=False)
    plt.title(f"Boxplot de {column}")
    plt.xlabel(column)
    plt.show()
```

# PASO 2: Análisis Bivariado

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Variables categóricas a analizar
categorical_columns = ['gender', 'married', 'paymentmethod']

for column in categorical_columns:
  
    churn_counts = pd.crosstab(df[column], df['is_churner'], normalize='index')
    churn_counts.plot(kind='bar', stacked=True, figsize=(8, 5), colormap="viridis")
    plt.title(f"Distribución de {column} por churn")
    plt.xlabel(column)
    plt.ylabel("Proporción")
    plt.legend(["No Churn", "Churn"])
    plt.show()
```


Insight clave: Los métodos de pago automatizados parecen actuar como un factor protector contra el churn.
Acción sugerida: La empresa podría fomentar el uso de métodos automáticos mediante incentivos para mejorar la retención

```python
# Variables numéricas a analizar
numerical_columns = ['age', 'contract_duration_days', 'children']

for column in numerical_columns:
    sns.boxplot(data=df, x='is_churner', y=column)
    plt.title(f"{column} vs Churn")
    plt.xlabel("Churn (0 = No, 1 = Sí)")
    plt.ylabel(column)
    plt.show()
```

Conclusión 1
No parece haber una relación fuerte entre la edad (age) y el churn (is_churner).
Esto indica que el comportamiento de churn no depende directamente de la edad del cliente

Conclusión 2
Relación clara: Los churners tienden a tener contratos más cortos que los no churners.
Insight potencial: La duración del contrato podría ser una métrica importante para predecir el churn.

Conclusión 3
Relación débil pero notable: Los churners suelen ser clientes sin hijos, mientras que los no churners incluyen una proporción mayor de clientes con hijos.
Este patrón podría sugerir que los clientes con responsabilidades familiares son menos propensos a abandonar el servicio.

#### Validación de Consistencia en los Datos


#### **Pasos Realizados**

1. **Revisión de duplicados y valores nulos:**
   - **Duplicados:**
     - Resultado: **0 duplicados encontrados.**
   - **Valores nulos:**
     - Decidimos **eliminar los registros con valores nulos**, dado que imputarlos podría comprometer la calidad de los datos.

2. **Validación de fechas:**
   - Confirmamos que la columna `billing_date` tiene un rango válido entre enero y octubre de 2023.
   - Detectamos facturación **irregular** en algunos clientes, lo que sugiere períodos no consecutivos de facturación.

3. **Relación entre servicios y cargos mensuales:**
   - Creamos una nueva columna llamada `total_services` para contar los servicios activos por cliente.


4. **Identificación y manejo de registros inconsistentes:**
   - Creamos una bandera (`is_inconsistent`) para marcar registros problemáticos.







```python
# Verificar el rango de fechas en billing_date
print(f"Rango de fechas: {df2_cleaned['billing_date'].min()} a {df2_cleaned['billing_date'].max()}")

# Analizar periodos de facturación por cliente
facturacion_inconsistente = df2_cleaned.groupby('customerid')['billing_date'].apply(
    lambda x: (x.diff().dt.days.dropna() != 30).sum()
)
print("\nClientes con periodos de facturación irregulares:")
print(facturacion_inconsistente[facturacion_inconsistente > 0])
```

```python
# Crear una bandera para identificar clientes con facturación irregular
df2_cleaned['is_irregular'] = df2_cleaned['customerid'].isin(facturacion_inconsistente.index)

# Verificar proporción de clientes irregulares
proporcion_irregular = df2_cleaned['is_irregular'].mean()
print(f"Proporción de clientes con facturación irregular: {proporcion_irregular:.2%}")

# Continuar con la relación entre cargos y servicios
servicios = ['onlinesecurity', 'onlinebackup', 'deviceprotection',
             'techsupport', 'streaming']
df2_cleaned['total_services'] = df2_cleaned[servicios].apply(
    lambda row: sum(row == 'Yes'), axis=1
)

# Analizar la relación entre total_services y monthlycharges
sns.boxplot(x='total_services', y='monthlycharges', data=df2_cleaned)
plt.title("Relación entre número de servicios activos y cargos mensuales")
plt.xlabel("Número de servicios activos")
plt.ylabel("Cargos mensuales")
plt.show()
```

```python
# Crear una bandera para inconsistencias
df2_cleaned['is_inconsistent'] = False

# Marcar clientes con 0 servicios activos y cargos altos
df2_cleaned.loc[(df2_cleaned['total_services'] == 0) &
                (df2_cleaned['monthlycharges'] > 50), 'is_inconsistent'] = True

# Marcar clientes con 5 servicios activos y cargos bajos
df2_cleaned.loc[(df2_cleaned['total_services'] == 5) &
                (df2_cleaned['monthlycharges'] < 20), 'is_inconsistent'] = True

inconsistencias_resumen = df2_cleaned['is_inconsistent'].value_counts()
print("Resumen de registros inconsistentes:")
print(inconsistencias_resumen)
```

```python
# Filtrar registros inconsistentes
registros_inconsistentes = df2_cleaned[df2_cleaned['is_inconsistent']]
print("Registros inconsistentes:")
print(registros_inconsistentes.head())


df2_cleaned = df2_cleaned[~df2_cleaned['is_inconsistent']]
```

```python
# Guardar los registros inconsistentes para análisis posterior
registros_inconsistentes.to_csv("registros_inconsistentes.csv", index=False)

# Eliminar registros inconsistentes del DataFrame principal
df2_cleaned = df2_cleaned[~df2_cleaned['is_inconsistent']]


print(f"Tamaño del DataFrame después de eliminar inconsistencias: {df2_cleaned.shape}")
```

CREAR CAMPAÑA DE RETENCION DE CLIENTES:

```python
# Identificar clientes en riesgo sociodemográfico (Pregunta 1)
high_risk_socdem = df_cleaned[
    (df_cleaned['gender'] == churners['gender'].mode()[0]) &
    (df_cleaned['age'] >= churners['age'].mean() - churners['age'].std()) &
    (df_cleaned['age'] <= churners['age'].mean() + churners['age'].std()) &
    (df_cleaned['married'] == churners['married'].mode()[0]) &
    (df_cleaned['is_churner'] == 0)  # Clientes activos
    ]

# Agregar criterios económicos y de comportamiento (Pregunta 3)
high_risk_combined = high_risk_socdem[
    (high_risk_socdem['monthlycharges'] > churners['monthlycharges'].mean()) &
    (high_risk_socdem['service_combination'].isin(service_comb_dist.index)) &
    (high_risk_socdem['paymentmethod'] == 'Electronic check')  # Método de pago problemático
    ]

# Número total de clientes en riesgo identificados
num_high_risk_combined = high_risk_combined.shape[0]
print(f"Número de clientes en riesgo identificados: {num_high_risk_combined}")

# Simulación de la campaña
# Costos de retención y tasas de éxito
retention_cost = 10  # Costo por cliente
optimistic_success_rate = 0.8
conservative_success_rate = 0.5

# Ingresos retenidos y costos
avg_monthly_income = high_risk_combined['monthlycharges'].mean()
avg_tenure_remaining = high_risk_combined['remaining_tenure'].mean()

# Escenario optimista
opt_revenue_retained = num_high_risk_combined * avg_monthly_income * avg_tenure_remaining * optimistic_success_rate
opt_campaign_cost = num_high_risk_combined * retention_cost
opt_net_gain = opt_revenue_retained - opt_campaign_cost

# Escenario conservador
cons_revenue_retained = num_high_risk_combined * avg_monthly_income * avg_tenure_remaining * conservative_success_rate
cons_campaign_cost = num_high_risk_combined * retention_cost
cons_net_gain = cons_revenue_retained - cons_campaign_cost

print(f"Escenario Optimista: Ingresos Retenidos: {opt_revenue_retained:.2f}, Costo de la Campaña: {opt_campaign_cost:.2f}, Ganancia Neta: {opt_net_gain:.2f}")
print(f"Escenario Conservador: Ingresos Retenidos: {cons_revenue_retained:.2f}, Costo de la Campaña: {cons_campaign_cost:.2f}, Ganancia Neta: {cons_net_gain:.2f}")
```
